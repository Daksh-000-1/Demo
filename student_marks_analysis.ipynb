Only using Basic python to sove the the problem of a teacher to analysis a students marks.

```python
data = {"Aman":78, "Riya":56, "Karan":89, "Neha":90, "Arjun":45, "Priya":67, "Rahul":72, "Simran":88, "Vikas":59, "Anjali":91}
marks = data.values()
print("Average Marks: ",sum(marks)/len(marks))
print("Highest Marks: ",max(marks))
print("Lowest Marks: ",min(marks))
c=0
c2=0
passed_student =[]
for i in marks:
  if i>75:
    c+=1
  if i<60:
    c2+=1
  if i >= 60:
    passed_student.append(i)
print("Above 75:", c ,"students")
print("Below 60:", c2 ,"students")
print("Passed Student: ",passed_student)
```
